### About this notebook 
The code below is used to anonymize data. 
Instead of hashing, we simply use the simplest method to analymize the data. 
We store a table in a MS Excel file named "AnanyKey", in which all our student logins were assigned a unique randomized number. We then substitute the user id with this randomized number


```python
import pandas as pd
# this is included in the script below
def anonymize_student(filename):
    import pandas as pd 
    import csv 
    
    df=pd.read_csv(filename)
    studentdf=pd.read_excel("data/AnanyKey.xlsx")
    studentdf['Usr_id']=studentdf['Usr_id'].astype(str)
    df=pd.merge(df,studentdf,left_on="user_id",right_on="USERNAME",how="left")
    df.drop(['user_id','USERNAME',"ACAD_PROG_DESCR","PROGRAM_YEAR","Column1"],axis=1)
    df=df[['Usr_id','date_time','url','domain','acad_prog','prog_yr','on_or_off']]
    return df
```

```python
import pandas as pd
import glob
import os

result=pd.DataFrame()
for filename in glob.iglob(os.path.join("combined/2017-18/","*.csv")):
    filename = filename.replace("\\","/")
    df=pd.read_csv(filename)
    studentdf=pd.read_excel("data/AnanyKey.xlsx")
    studentdf['Usr_id']=studentdf['Usr_id'].astype(str)
    df=pd.merge(df,studentdf,left_on="user_id",right_on="USERNAME",how="left")
    df.drop(['user_id','USERNAME',"ACAD_PROG_DESCR","PROGRAM_YEAR","Column1"],axis=1)
    df=df[['Usr_id','date_time','url','domain','acad_prog','prog_yr','on_or_off']]
    result=result.append(df)
    
result.to_csv("combined_ana_2017-18.csv")
```
